## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(data=df, x='ConvertedComp')
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q3 = df['ConvertedComp'].quantile(0.75)
q1 = df['ConvertedComp'].quantile(0.25)
IQR = q3 - q1
print('IQR:', IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
print('Upper bounds:', q3 + IQR*1.5)
print('Lower bounds:', q1 - IQR*1.5)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
up_bound = q3 + IQR*1.5
lw_bound = q1 - IQR*1.5

num_outliers = 0
for i in df['ConvertedComp']:
    if (i > up_bound) or (i < lw_bound):
        num_outliers = num_outliers + 1

num_outliers
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
outlier = df[(df['ConvertedComp'] > up_bound) | (df['ConvertedComp'] < lw_bound)].index
df_clean = df.drop(outlier)

print('before clean outlier:', df.shape[0])
print('after clean outlier:', df_clean.shape[0])
```

```python
df['ConvertedComp'].median()
```

```python
df_clean['ConvertedComp'].median()
```
